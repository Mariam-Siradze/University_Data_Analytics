**სავარჯიშო 1:** პითონში დაწერეთ 2 კამათლის 1000-ჯერ აგდების სასიმულაციო კოდი და გამოთვალეთ ერთდროულად 5-იანის და 2-იანის (ფანჯი-დუ) დაჯდომის ალბათობა

**0.3 ქულა**

````python
import random

def dice_roll():
    return random.randint(1,6)
    
const = 1000
count = 0  

data = [(dice_roll(), dice_roll()) for _ in range(const)]

for roll1, roll2 in data:
    if roll1 == 5 and roll2 == 2:
        count +=1

prob = count / const

print(prob)
````

**სავარჯიშო 2:** დააგენერირეთ შემთხვევითი ერთობლიობა (random sample) სამკუთხა განაწილებით და გაუკეთეთ ვიზუალიზება. გამოიყენეთ NumPy-ს დოკუმენტაცია: https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

````python
import numpy as np
import matplotlib.pyplot as plt

# პარამეტრები
left = -3
mode = 0
right = 8
size = 3000

# შევქმნათ შემთხვევითი ერთობლიობა სამკუთხა განაწილებით
samples = np.random.triangular(left, mode, right, size)

# ჰისტოგრამის ვიზუალიზება
plt.hist(samples, bins=30, density=True, alpha=0.5, color='b')

# Add labels and title
plt.xlabel('X')
plt.ylabel('ალბათობის სიმკვრივე')
plt.title('სამკუთხა განაწილება')

# გრაფიკის ჩვენება
plt.show()
````

**სავარჯიშო 3:** ელექტრონულ კომპონენტებს აწარმოებს A რობოტი და B რობოტი. A აწარმოებს კომპონენტების 60%-ს, ხოლო B დანარჩენ 40%-ს. ცნობილია, რომ წუნიანია (დეფექტურია) A-ს წარმოებულის 5%, ხოლო B-ს წარმოებულის 3 %. შემთხვევით ვირჩევთ დეფექტურ კომპონენტს და ვიგებთ, რომ A-ს უწარმოებია. რა არის ალბათობა იმისა, რომ მართლაც A-მ აწარმოა? დაწერეთ პითონის კოდი და კომენტარებში ახსენით.

````python
a_rob = 0.6
b_rob = 0.4

a_def_prob = 0.05
b_def_prob = 0.03

tot_def_prob = a_rob * a_def_prob + a_rob * b_def_prob

tot_prob_a = (a_rob * a_def_prob) / tot_def_prob

print('result', tot_prob_a)
````

**სავარჯიშო 4:** მოცემულია ფრეშეს განაწილება (იხ. ქვემოთ კოდის ბლოკი). გამოთვალეთ განაწილების საშუალო, დისპერსია, სტანდარტული გადახრა, ასიმეტრია (skewness), ექსცესი (kurtosis) და ჭარბი ექცესი (excess kurtosis).

**0.8 ქულა**

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# პარამეტრები
shape = 2.0
scale = 1.0

# შემთხვევითი ერთობლიობა ფრეშეს განაწილებით
np.random.seed(42)  # რეპროდუცირებისთვის
samples = stats.invweibull.rvs(shape, scale=scale, size=10000)
samples = samples[samples<=12]

# ჰისტოგრამა
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b', label='ჰისტოგრამა')

# განაწილების ფუნქცია
x = np.linspace(0, 5, 10000)
pdf = stats.invweibull.pdf(x, shape, scale=scale)
plt.plot(x, pdf, 'r-', lw=2, label='განაწილების ფუნქცია')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('სიმკვრივე')
plt.title('Frechet (Inverse Weibull) Distribution')
plt.legend()


# Show plot
plt.show()
````

````python
# განაწილების მახასიათებლების გამოთვლა
mean = np.mean(samples)
var = np.var(samples)
std = np.std(samples)
skew = stats.skew(samples)
kurt = stats.kurtosis(samples)
exc_kurt = kurt - 3

print('საშუალო', mean, '\nდისპერსია', var, '\nსტანდარტული გადახრა', std, '\nასიმეტრია', skew, '\nექსცესი', kurt, '\nჭარბი ექსცესი', exc_kurt)
````

**სავარჯიშო 5:** სტანდარტული გადახრის და ინტერკვარტილური დიაპაზონის მეთოდებით იპოვეთ ამოვარდნილი მნიშვნელობები (outliers) მოცემულ მასივში.

1. **სტანდარტული გადახრის მეთოდი**:
   - $\text{ზედა ზღვარი} = \text{საშუალო} + k \times \text{სტანდარტული გადახრა} $
   - $ \text{ქვედა ზღვარი} = \text{საშუალო} - k \times \text{სტანდარტული გადახრა} $

2. **ინტერკვარტილური დიაპაზონის (IQR) მეთოდი**:
   - გამოვთვლით მონაცემების პირველ (Q1) და მესამე კვარტილებს (Q3).
   - ვითვლით ინტერკვარტილურ დიაპაზონს: $ \text{IQR} = Q3 - Q1 $
   - $ \text{ზედა ზღვარი} = Q3 + k \times \text{IQR} $
   -$ \text{ქვედა ზღვარი} = Q1 - k \times \text{IQR} $

$ $
**2 ქულა**

````python
import numpy as np

# ნორმალურად განაწილებული მასივი
samples = np.random.normal(loc=0, scale=1, size=100)

# ამოვარნილი მნიშვნელობების დამატება მასივზე
outliers = np.random.normal(loc=10, scale=5, size=5)
samples_with_outliers = np.concatenate((samples, outliers))

# სტანდარტული გადახრის (Z-score) მეთოდის ფუნქცია
def detect_outliers_zscore(data, threshold=3):
    z_scores = (data - np.mean(data)) / np.std(data)
    return np.abs(z_scores) > threshold

# IQR მეთოდის ფუნქცია
def detect_outliers_iqr(data, threshold=3):
    quartile_1, quartile_3 = np.percentile(data, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - threshold * iqr
    upper_bound = quartile_3 + threshold * iqr
    return (data < lower_bound) | (data > upper_bound)


# გამოვთვალოთ ამოვარდნილი მნიშვნელობები სტანდარტული გადახრის და ინტერკვარტილური დიაპაზონის მეთოდებით
outliers_zscore = detect_outliers_zscore(samples_with_outliers)  # სტანდარტული გადახრის
outliers_iqr = outliers_iqr = detect_outliers_iqr(samples_with_outliers) # ინტერკვარტილური

print("სტანდარტული გადახრის მეთოდით:", outliers_zscore)
print("IQR მეთოდით:", outliers_zscore)
````
